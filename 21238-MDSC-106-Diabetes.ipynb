```python
# definig a function which displays all the below plots for all attributes in the dataframe
def diagnostic_plots(df, variable,target):
    plt.figure(figsize=(20, 7))

    # lineplot
    plt.subplot(1, 5, 1)
    plt.plot(df[variable],color='g')
    plt.title('Lineplot')
    
    # histogram
    plt.subplot(1, 5, 2)
    sns.histplot(df[variable], kde = True,color='g')
    plt.title('Histogram')


    # scatterplot
    plt.subplot(1, 5, 3)
    plt.scatter(df[variable],df[target],color = 'b')
    plt.title('Scatterplot')
    
    # boxplot
    plt.subplot(1, 5, 4)
    sns.boxplot(y=df[variable],color = 'r')
    plt.title('Boxplot')
    
    # barplot
    plt.subplot(1, 5, 5)
    sns.barplot(x = target, y = variable, data = df,color = 'g')   
    plt.title('Barplot')
    
    plt.show()

for col in df:
    diagnostic_plots(df,col,'Outcome')
```

```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
# giving names to all columns as attributes in the dataframe
attributes = ['Class','Alcohol','Malic Acid','Ash','Alcanlinity of ash','Magnesium','Total Phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color Intensity','Hue','OD280/OD315 of diluted wines','Proline']

# reading the dataset and printing it
df = pd.read_csv("datasets/Wine dataset.csv")

df.columns = attributes
df
```

```python
# Count plot for the first attribute - Class
sns.countplot(data=df,x='Class')
plt.ylabel('Count')
plt.title('Count plot for Class attribute')
plt.show()
```

```python
# Scatterplot between Alcohol and Flavanoids basing on their classes
sns.scatterplot(data=df,x='Alcohol',y='Flavanoids',hue='Class',palette=['red','blue','green'])
plt.title('Scatterplot for Alcohol and Flavanoids based on their classes')
plt.show()
```

```python
# Grid plot for Magnesium vs all other attributes basing on their Classes using scatterplot
fig, axes = plt.subplots(4,3,figsize=(20,20)) # required rows,cols and plotsize
fig.subplots_adjust(hspace=0.3, wspace=0.3) # required height and width space between two sub plots

# for loop to compare Magnesium with all the other attributes basing on their Classes
i=j=0
for feature in df.columns:
    if j > 2:
        j = 0
        i += 1
    
    if feature != 'Magnesium' and feature != 'Class':  # if statement which doesn't conisder magnesium or class as a value
        sns.scatterplot(data=df,x='Magnesium',y=feature,hue='Class',ax=axes[i,j],palette=['red','blue','green'])
        axes[i,j].set_title('Magnesium vs {} '. format(feature))
        j +=1
```
